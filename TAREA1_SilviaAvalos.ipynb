### Graficar los valores globales y los de M√©xico para: 'Total', 'Coal', 'Oil', 'Gas', 'Cement', 'Flaring', 'Per Capita'.

```python
#Del conjunto de datos completo, obtenemos todos los valores globales y los de M√©xico
globald = df[(df['Country'] == 'Global')]
#globald
mexicod = df[(df['Country'] == 'Mexico')]
#mexicod

#Hacemos una lista con los valores que se nos pide graficar
valoresgraf = ['Total', 'Coal', 'Oil', 'Gas', 'Cement', 'Flaring', 'Per Capita']

#Con un ciclo for recorremos y graficamos los datos de la lista hecha anteriormente, se grafican los datos globales vs los de M√©xico
for valoresgraf in valoresgraf:
    plt.title(valoresgraf)
    plt.xlabel('A√±o')
    plt.ylabel('MtCO2')
    plt.plot(globald['Year'], globald[valoresgraf], label='Global')
    plt.plot(mexicod['Year'], mexicod[valoresgraf], label='M√©xico')
    plt.legend()
    plt.show()
```

### Extraer datos de los pa√≠ses pertenecientes al G20 y graficar las columnas del punto anterior.

### - Considerando solo la emisi√≥n de los paises del G20, ¬øCon qu√© porcentaje contribuye M√©xico al Total y Per Capita?



```python
#Del conjunto de datos completo, obtenemos todos los valores globales y los de M√©xico
arg = df[(df['Country'] == 'Argentina')]
aust = df[(df['Country'] == 'Australia')]
bra = df[(df['Country'] == 'Brazil')]
can = df[(df['Country'] == 'Canada')]
chin = df[(df['Country'] == 'China')]
fran = df[(df['Country'] == 'France')]
indi = df[(df['Country'] == 'India')]
indo = df[(df['Country'] == 'Indonesia')]
ita = df[(df['Country'] == 'Italy')]
jap = df[(df['Country'] == 'Japan')]
mex = df[(df['Country'] == 'Mexico')]
kor = df[(df['Country'] == 'South Korea')]
rus = df[(df['Country'] == 'Russia')]
safr = df[(df['Country'] == 'South Africa')]
tur = df[(df['Country'] == 'Turkey')]
ukin = df[(df['Country'] == 'United Kingdom')]
usa = df[(df['Country'] == 'USA')]
sarab = df[(df['Country'] == 'Saudi Arabia SAU')]
germ = df[(df['Country'] == 'Germany')]

#Usamos la lista de valores creada en el punto anterior "valoresgraf"
valoresgraf = ['Total', 'Coal', 'Oil', 'Gas', 'Cement', 'Flaring', 'Per Capita']
#Con un ciclo for recorremos y graficamos los datos de la lista hecha anteriormente, se grafican los datos de la G20
for valoresgraf in valoresgraf:
    plt.title(valoresgraf)
    plt.xlabel('A√±o')
    plt.ylabel('MtCO2')
    plt.plot(arg['Year'], arg[valoresgraf], label='Argentina')
    plt.plot(aust['Year'], aust[valoresgraf], label='Australia')
    plt.plot(bra['Year'], bra[valoresgraf], label='Brasil')
    plt.plot(can['Year'], can[valoresgraf], label='Canad√°')
    plt.plot(chin['Year'], chin[valoresgraf], label='China')
    plt.plot(fran['Year'], fran[valoresgraf], label='Francia')
    plt.plot(indi['Year'], indi[valoresgraf], label='India')
    plt.plot(indo['Year'], indo[valoresgraf], label='Indonesia')
    plt.plot(ita['Year'], ita[valoresgraf], label='Italia')
    plt.plot(jap['Year'], jap[valoresgraf], label='Japon')
    plt.legend()
    plt.show()
```

```python
#Usamos la lista de valores creada en el punto anterior "valoresgraf"
valoresgraf = ['Total', 'Coal', 'Oil', 'Gas', 'Cement', 'Flaring', 'Per Capita']
#Con un ciclo for recorremos y graficamos los datos de la lista hecha anteriormente, se grafican los datos de la G20
for valoresgraf in valoresgraf:
    plt.title(valoresgraf)
    plt.xlabel('A√±o')
    plt.ylabel('MtCO2')
    plt.plot(mex['Year'], mex[valoresgraf], label='M√©xico')
    plt.plot(kor['Year'], kor[valoresgraf], label='Sur Corea')
    plt.plot(rus['Year'], rus[valoresgraf], label='Rusia')
    plt.plot(safr['Year'], safr[valoresgraf], label='Sud√°frica')
    plt.plot(tur['Year'], tur[valoresgraf], label='Turqu√≠a')
    plt.plot(ukin['Year'], ukin[valoresgraf], label='Reino Unido')
    plt.plot(usa['Year'], usa[valoresgraf], label='USA')
    plt.plot(sarab['Year'], sarab[valoresgraf], label='Arabia Saudita')
    plt.plot(germ['Year'], germ[valoresgraf], label='Alemania')
    plt.legend()
    plt.show()
```

```python
#Para contestar las siguientes preguntas, creamos un nuevo dataframe que solo incluya a los pa√≠ses pertenecientes a la G20
G20 = pd.concat([arg,aust,bra,can,chin,fran,indi,indo,ita,jap,mex,kor,rus,safr,tur,ukin,usa,sarab,germ], axis=0)
#G20

#¬øCu√°les son los 3 pa√≠ses que m√°s contaminan en cada uno de los casos?
suma = G20.groupby('Country').sum()
#suma
pais_mas_contamina = {}
valoresgraf = ['Total', 'Coal', 'Oil', 'Gas', 'Cement', 'Flaring', 'Per Capita']

for i in valoresgraf:
  pais_mas_contamina.update({i : suma.reset_index().nlargest(3, i)['Country'].tolist()})
print(f'Los 3 pa√≠ses que m√°s contaminan en cada caso son: \n{pais_mas_contamina}\n')

#Considerando solo la emisi√≥n de los paises del G20, ¬øCon qu√© porcentaje contribuye M√©xico al Total y Per Capita?
#An√°lisis por columna del total de M√©xico y la suma de todos los pa√≠ses del G20
colmexico = {}
for i in valoresgraf:
    totmexico = suma.loc['Mexico', i]
    totglobal = suma[i].sum()
    porcentage = totmexico / totglobal * 100
    colmexico.update({i : porcentage})
    print(f"De los pa√≠ses pertenecientes a la G20, M√©xico representa el {colmexico[i]:.2f}% del valor total de {i}")
```
