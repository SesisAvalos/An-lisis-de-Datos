```python
import matplotlib.pyplot as plt
import numpy as np
import random


plt.figure(figsize=(9,9))           #Creamos una figura de 9x9

#A partir de aquí mostramos rectángulos de diferentes tamaños que van dentro de la imagen anterior
#Y cada retángulo lo rellenamos con algo

#Rect 1
subf1 = plt.subplot(4,4,(1,3))      
np.random.seed(19680801)
N = 50
x1 = np.random.rand(N)
y1 = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2  
subf1.scatter(x1, y1, s=area, c=colors, alpha=0.5)

#Rect 2
subf2 = plt.subplot(4,4,(5,6))
np.random.seed(19680801)
N2 = 100
r0 = 0.6
x2 = 0.9 * np.random.rand(N2)
y2 = 0.9 * np.random.rand(N2)
area = (20 * np.random.rand(N2))**2  # 0 to 10 point radii
c = np.sqrt(area)
r = np.sqrt(x2 ** 2 + y2 ** 2)
area1 = np.ma.masked_where(r < r0, area)
area2 = np.ma.masked_where(r >= r0, area)
subf2.scatter(x2, y2, s=area1, marker='^', c=c)
subf2.scatter(x2, y2, s=area2, marker='o', c=c)
# Show the boundary between the regions:
theta = np.arange(0, np.pi / 2, 0.01)
subf2.plot(r0 * np.cos(theta), r0 * np.sin(theta))


#Rect 3
subf3 = plt.subplot(4,4,9)
n_buses = 50
bus_times = np.asarray([random.randint(0, 59) for _ in range(n_buses)])
bus_likelihood = np.asarray([random.random() for _ in range(n_buses)])
subf3.scatter(x=bus_times, y=bus_likelihood, c=colors)


#Rect 4
subf4 = plt.subplot(4,4,10)
x4 = np.linspace(0, 10, 30)
y4 = np.sin(x4)

subf4.plot(x4, y4, 'o', color='black');


#Rect 5
subf5 = plt.subplot(4,4,(7,11))
# dataset-1
x11 = [89, 43, 36, 36, 95, 10,
      66, 34, 38, 20]
 
y11 = [21, 46, 3, 35, 67, 95,
      53, 72, 58, 10]
# dataset2
x2 = [26, 29, 48, 64, 6, 5,
      36, 66, 72, 40]
 
y2 = [26, 34, 90, 33, 38,
      20, 56, 2, 47, 15]
 
subf5.scatter(x11, y11, c ="pink",
            linewidths = 2,
            marker ="s",
            edgecolor ="blue",
            s = 50)
 
subf5.scatter(x2, y2, c ="black",
            linewidths = 2,
            marker ="^",
            edgecolor ="red",
            s = 200)
```
