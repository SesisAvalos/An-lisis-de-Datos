```python
#Convertir a blanco 
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from skimage.transform import resize

#Este código fue realizado con la ayuda de un compañero de clase
imagen = Image.open('/content/WhatsApp Image 2023-02-27 at 2.49.49 PM.jpeg') 
ima = imagen.resize((360,450))
ima.save("ProyectosconPython_resize.png")
ima
```

```python
im1 = np.array(ima)
for i in range(im1.shape[0]):
  for j in range(im1.shape[1]):
    a =  max(im1[i][j][0], im1[i][j][1], im1[i][j][2])
    im1[i][j] = [a,a,a]

fig,ax = plt.subplots(figsize=(6,6))
ax.imshow(im1)
ax.set_xticks([])
ax.set_yticks([]);
```

```python
from scipy.ndimage import gaussian_filter

smooth = gaussian_filter(ima, 5, order=0, mode='nearest')

fig,ax = plt.subplots(figsize=(6,6))
ax.imshow(smooth)
ax.set_xticks([])
ax.set_yticks([]);
```
