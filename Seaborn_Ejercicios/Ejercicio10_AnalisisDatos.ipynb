```python
import seaborn as sns
sns.get_dataset_names()
```

```python
df = sns.load_dataset('iris')
df.head()
```

### Displot

```python
sns.displot(df['sepal_length'])
```

### Joinplot

```python
sns.jointplot(data = df, x = "petal_length", y = "petal_width", hue="species",palette="cool")
```

### Pairplot

```python
sns.pairplot(df, hue="species", height=1.5, palette='plasma');
```
